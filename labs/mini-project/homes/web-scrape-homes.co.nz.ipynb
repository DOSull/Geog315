```python
import requests

import pandas
import json
import time
```

```python
homes_list = pandas.read_csv('homes.csv')
urls = list(homes_list.url)
urls
```

```python
records = []

for url in urls:
    home = requests.get('https://gateway.homes.co.nz/property', params={'url': url})
    print(home.url)
    prop_info = json.loads(home.text)['card']['property_details']
    prop_info.update(json.loads(home.text)['card']['point'])
    prop_info['price'] = json.loads(home.text)['card']['display_price']
    records.append(prop_info)
    time.sleep(1)
```

```python
homes_data = pandas.DataFrame(records)[['address', 'num_bathrooms', 'num_bedrooms', 'lat', 'long', 'price']]
homes_data.head()
```

```python
homes_data.to_csv('homes-data.csv')
```
